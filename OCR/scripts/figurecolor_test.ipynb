```python
# ============================================================
# PaddleOCR + Lab色空間による文字色抽出・マスク確認コード
# 修正版：PaddlePaddle 3.2.2固定 / cls=True不使用
# 対象: Google Drive / マイドライブ / 評価用3枚 にある画像3枚
# ============================================================

# ============================================================
# 1. ライブラリを入れ直す
# ============================================================

!pip -q uninstall -y paddlepaddle paddleocr paddlex

!pip -q install paddlepaddle==3.2.2 -i https://www.paddlepaddle.org.cn/packages/stable/cpu/
!pip -q install paddleocr
!pip -q install opencv-python-headless matplotlib numpy
!pip -q install langchain-text-splitters langchain-community langchain-core

# ============================================================
# 2. Google Driveをマウント
# ============================================================

from google.colab import drive
drive.mount('/content/drive')

# ============================================================
# 3. ライブラリ読み込み
# ============================================================

import os
import glob
import cv2
import numpy as np
import matplotlib.pyplot as plt
import paddle
from paddleocr import PaddleOCR

print("PaddlePaddle version:", paddle.__version__)

# ============================================================
# 4. 設定
# ============================================================

IMAGE_DIR = "/content/drive/MyDrive/評価用3枚"
OUTPUT_DIR = os.path.join(IMAGE_DIR, "result_lab_color_mask")
TEMP_DIR = os.path.join(IMAGE_DIR, "temp_resized_for_ocr")

os.makedirs(OUTPUT_DIR, exist_ok=True)
os.makedirs(TEMP_DIR, exist_ok=True)

MAX_TARGETS = 3

# OCRの最低信頼度
MIN_OCR_SCORE = 0.40

# 画像を大きすぎないサイズに縮小する
MAX_SIDE_FOR_OCR = 2400

# Lab色空間で「文字色に近い」とみなす距離
LAB_COLOR_DISTANCE_THRESHOLD = 28

print("=== 設定確認 ===")
print("画像フォルダ:", IMAGE_DIR)
print("結果保存先:", OUTPUT_DIR)
print("一時画像保存先:", TEMP_DIR)
print("最大対象文字列数:", MAX_TARGETS)
print("OCR最低信頼度:", MIN_OCR_SCORE)
print("OCR用最大辺:", MAX_SIDE_FOR_OCR)
print("Lab色距離しきい値:", LAB_COLOR_DISTANCE_THRESHOLD)

# ============================================================
# 5. PaddleOCR初期化
# ============================================================

# 今回は文字検出・文字認識を見たいので、
# 文書の向き補正やゆがみ補正は切って軽くする
ocr = PaddleOCR(
    lang="en",
    use_doc_orientation_classify=False,
    use_doc_unwarping=False,
    use_textline_orientation=False
)

print("PaddleOCR 初期化 OK")

# ============================================================
# 6. 補助関数
# ============================================================

def show_image(title, img_rgb, figsize=(8, 5)):
    plt.figure(figsize=figsize)
    plt.imshow(img_rgb)
    plt.title(title)
    plt.axis("off")
    plt.show()


def resize_for_ocr(image_path, max_side=2400):
    """
    大きすぎる画像をOCRしやすいサイズに縮小する。
    戻り値:
      resized_path, resized_bgr, scale
    scaleは「縮小後 / 元画像」の倍率。
    """
    image_bgr = cv2.imread(image_path)

    if image_bgr is None:
        raise FileNotFoundError(f"画像を読み込めません: {image_path}")

    h, w = image_bgr.shape[:2]
    long_side = max(h, w)

    if long_side <= max_side:
        scale = 1.0
        resized_bgr = image_bgr.copy()
    else:
        scale = max_side / long_side
        new_w = int(w * scale)
        new_h = int(h * scale)
        resized_bgr = cv2.resize(image_bgr, (new_w, new_h), interpolation=cv2.INTER_AREA)

    base = os.path.splitext(os.path.basename(image_path))[0]
    resized_path = os.path.join(TEMP_DIR, f"{base}_resized.jpg")
    cv2.imwrite(resized_path, resized_bgr)

    return resized_path, resized_bgr, scale


def normalize_box(box):
    """
    OCR結果のboxを [[x,y], [x,y], [x,y], [x,y]] にそろえる。
    """
    box = np.array(box)

    # [x1, y1, x2, y2] の場合
    if box.ndim == 1 and len(box) == 4:
        x1, y1, x2, y2 = box.astype(int)
        box = np.array([
            [x1, y1],
            [x2, y1],
            [x2, y2],
            [x1, y2]
        ])

    return box.astype(int)


def parse_paddleocr_result(raw_result):
    """
    PaddleOCR v3系の結果を、
    [{'text':..., 'score':..., 'box':...}, ...]
    に整える。
    """
    parsed = []

    if raw_result is None:
        return parsed

    try:
        for res in raw_result:
            # dictのように扱える場合
            if hasattr(res, "get"):
                texts = res.get("rec_texts", [])
                scores = res.get("rec_scores", [])

                boxes = None
                for key in ["rec_polys", "dt_polys", "rec_boxes", "dt_boxes"]:
                    if key in res:
                        boxes = res.get(key)
                        break

                if boxes is None:
                    continue

                for text, score, box in zip(texts, scores, boxes):
                    if text is None:
                        continue

                    parsed.append({
                        "text": str(text),
                        "score": float(score),
                        "box": normalize_box(box)
                    })

            # 万が一、json形式で持っている場合
            elif hasattr(res, "json"):
                data = res.json

                if isinstance(data, dict):
                    texts = data.get("rec_texts", [])
                    scores = data.get("rec_scores", [])

                    boxes = None
                    for key in ["rec_polys", "dt_polys", "rec_boxes", "dt_boxes"]:
                        if key in data:
                            boxes = data.get(key)
                            break

                    if boxes is None:
                        continue

                    for text, score, box in zip(texts, scores, boxes):
                        parsed.append({
                            "text": str(text),
                            "score": float(score),
                            "box": normalize_box(box)
                        })

    except Exception as e:
        print("OCR結果の解析で問題:", e)

    return parsed


def run_ocr(image_path):
    """
    PaddleOCR v3のpredictのみを使う。
    cls=Trueは使わない。
    """
    raw = ocr.predict(input=image_path)
    parsed = parse_paddleocr_result(raw)
    return parsed


def crop_by_box(image_bgr, box, padding=10):
    """
    OCRで検出した文字領域を少し余裕を持って切り出す。
    """
    h, w = image_bgr.shape[:2]

    x_min = max(int(np.min(box[:, 0])) - padding, 0)
    y_min = max(int(np.min(box[:, 1])) - padding, 0)
    x_max = min(int(np.max(box[:, 0])) + padding, w - 1)
    y_max = min(int(np.max(box[:, 1])) + padding, h - 1)

    crop = image_bgr[y_min:y_max, x_min:x_max].copy()

    local_box = box.copy()
    local_box[:, 0] -= x_min
    local_box[:, 1] -= y_min

    return crop, local_box, (x_min, y_min, x_max, y_max)


def create_polygon_mask(crop_bgr, local_box):
    """
    OCRが文字と判断した四角形領域のマスクを作る。
    白い部分が処理対象。
    """
    mask = np.zeros(crop_bgr.shape[:2], dtype=np.uint8)
    cv2.fillPoly(mask, [local_box.astype(np.int32)], 255)
    return mask


def extract_text_color_lab(crop_bgr, polygon_mask):
    """
    OCR領域内から、文字っぽい色をLab色空間で推定する。

    考え方:
    1. OCR領域内の代表的な色を背景色とみなす
    2. 背景色から離れている画素を文字っぽい画素とみなす
    3. その画素の代表Lab値を文字色として扱う
    """
    lab = cv2.cvtColor(crop_bgr, cv2.COLOR_BGR2LAB)

    inside_pixels = lab[polygon_mask > 0]

    if len(inside_pixels) == 0:
        return None, None, None, None

    background_lab = np.median(inside_pixels, axis=0)

    diff_from_bg = np.linalg.norm(
        lab.astype(np.float32) - background_lab.astype(np.float32),
        axis=2
    )

    inside_diff = diff_from_bg[polygon_mask > 0]

    threshold_bg = np.percentile(inside_diff, 75)
    threshold_bg = max(threshold_bg, 10)

    text_like_mask = np.zeros(crop_bgr.shape[:2], dtype=np.uint8)
    text_like_mask[(diff_from_bg > threshold_bg) & (polygon_mask > 0)] = 255

    selected_count = np.sum(text_like_mask > 0)
    area_count = np.sum(polygon_mask > 0)

    # 文字っぽい画素が少なすぎる場合は、少し条件をゆるめる
    if area_count > 0 and selected_count < area_count * 0.01:
        threshold_bg = np.percentile(inside_diff, 60)
        text_like_mask[:] = 0
        text_like_mask[(diff_from_bg > threshold_bg) & (polygon_mask > 0)] = 255

    kernel = np.ones((2, 2), np.uint8)
    text_like_mask = cv2.morphologyEx(text_like_mask, cv2.MORPH_CLOSE, kernel)

    text_pixels = lab[text_like_mask > 0]

    if len(text_pixels) == 0:
        return None, background_lab, text_like_mask, None

    text_color_lab = np.median(text_pixels, axis=0)

    diff_from_text = np.linalg.norm(
        lab.astype(np.float32) - text_color_lab.astype(np.float32),
        axis=2
    )

    color_mask = np.zeros(crop_bgr.shape[:2], dtype=np.uint8)
    color_mask[(diff_from_text < LAB_COLOR_DISTANCE_THRESHOLD) & (polygon_mask > 0)] = 255
    color_mask = cv2.morphologyEx(color_mask, cv2.MORPH_CLOSE, kernel)

    return text_color_lab, background_lab, text_like_mask, color_mask


def apply_mask_to_crop(crop_bgr, mask):
    """
    マスクされた部分だけを残し、それ以外を白にする。
    """
    result = np.ones_like(crop_bgr) * 255
    result[mask > 0] = crop_bgr[mask > 0]
    return result


def draw_ocr_boxes(image_bgr, results):
    """
    元画像にOCR結果の枠と番号を描く。
    """
    vis = image_bgr.copy()

    for i, r in enumerate(results, start=1):
        box = r["box"].astype(np.int32)
        cv2.polylines(vis, [box], True, (0, 255, 0), 3)

        x = int(np.min(box[:, 0]))
        y = int(np.min(box[:, 1]))

        label = f"{i}: {r['text']} ({r['score']:.2f})"

        cv2.putText(
            vis,
            label,
            (x, max(y - 8, 25)),
            cv2.FONT_HERSHEY_SIMPLEX,
            0.7,
            (0, 255, 0),
            2,
            cv2.LINE_AA
        )

    return vis


# ============================================================
# 7. 画像取得
# ============================================================

image_paths = []

for ext in ["*.jpg", "*.jpeg", "*.JPG", "*.JPEG", "*.png", "*.PNG"]:
    image_paths.extend(glob.glob(os.path.join(IMAGE_DIR, ext)))

image_paths = sorted(image_paths)

print("\n=== 対象画像 ===")
for p in image_paths:
    print(os.path.basename(p))

if len(image_paths) == 0:
    raise FileNotFoundError("指定フォルダに画像が見つかりません。/content/drive/MyDrive/評価用3枚 を確認してください。")

image_paths = image_paths[:3]

# ============================================================
# 8. メイン処理
# ============================================================

all_results = []

for image_index, original_image_path in enumerate(image_paths, start=1):

    print("\n" + "=" * 80)
    print(f"画像 {image_index}: {os.path.basename(original_image_path)}")
    print("=" * 80)

    # 画像をOCR用に縮小
    ocr_image_path, image_bgr, scale = resize_for_ocr(
        original_image_path,
        max_side=MAX_SIDE_FOR_OCR
    )

    image_rgb = cv2.cvtColor(image_bgr, cv2.COLOR_BGR2RGB)

    print("ステータス1: 画像読み込み・縮小 OK")
    print("OCR用画像サイズ:", image_bgr.shape)
    print("縮小倍率:", round(scale, 3))
    print("OCR用画像:", ocr_image_path)

    # OCR実行
    try:
        ocr_results = run_ocr(ocr_image_path)
        print("ステータス2: PaddleOCR実行 OK")
    except Exception as e:
        print("ステータス2: PaddleOCR実行 NG")
        print("エラー内容:", e)
        continue

    print("検出された文字列数:", len(ocr_results))

    # 信頼度で絞る
    filtered_results = [
        r for r in ocr_results
        if r["score"] >= MIN_OCR_SCORE and r["text"].strip() != ""
    ]

    filtered_results = sorted(filtered_results, key=lambda x: x["score"], reverse=True)
    target_results = filtered_results[:MAX_TARGETS]

    print("ステータス3: 信頼度で絞り込み OK")
    print("今回処理する文字列数:", len(target_results))

    if len(target_results) == 0:
        print("この画像では処理対象となる文字列がありませんでした。")
        show_image("Original Image", image_rgb)
        continue

    # OCR検出結果を可視化
    boxed_bgr = draw_ocr_boxes(image_bgr, target_results)
    boxed_rgb = cv2.cvtColor(boxed_bgr, cv2.COLOR_BGR2RGB)

    show_image(
        f"OCR検出結果: {os.path.basename(original_image_path)}",
        boxed_rgb,
        figsize=(10, 6)
    )

    boxed_save_path = os.path.join(
        OUTPUT_DIR,
        f"{os.path.splitext(os.path.basename(original_image_path))[0]}_ocr_boxes.jpg"
    )
    cv2.imwrite(boxed_save_path, boxed_bgr)

    image_result = {
        "image_name": os.path.basename(original_image_path),
        "results": []
    }

    # 各OCR領域に対してLab色抽出
    for target_index, r in enumerate(target_results, start=1):

        text = r["text"]
        score = r["score"]
        box = r["box"]

        print("\n--- 対象", target_index, "---")
        print("OCR文字列:", text)
        print("OCR信頼度:", round(score, 3))

        crop_bgr, local_box, crop_rect = crop_by_box(image_bgr, box, padding=10)
        polygon_mask = create_polygon_mask(crop_bgr, local_box)

        text_color_lab, background_lab, text_like_mask, color_mask = extract_text_color_lab(
            crop_bgr,
            polygon_mask
        )

        if text_color_lab is None or color_mask is None:
            print("ステータス4: 文字色抽出 NG")
            print("理由: 文字っぽい画素を十分に取得できませんでした。")
            continue

        masked_crop_bgr = apply_mask_to_crop(crop_bgr, color_mask)

        crop_rgb = cv2.cvtColor(crop_bgr, cv2.COLOR_BGR2RGB)
        text_like_rgb = cv2.cvtColor(apply_mask_to_crop(crop_bgr, text_like_mask), cv2.COLOR_BGR2RGB)
        masked_crop_rgb = cv2.cvtColor(masked_crop_bgr, cv2.COLOR_BGR2RGB)

        print("ステータス4: Lab色空間で文字色抽出 OK")
        print("背景色の代表Lab:", np.round(background_lab, 1).tolist())
        print("文字色の代表Lab:", np.round(text_color_lab, 1).tolist())
        print("文字っぽい画素数:", int(np.sum(text_like_mask > 0)))
        print("文字色に近い画素数:", int(np.sum(color_mask > 0)))

        base_name = os.path.splitext(os.path.basename(original_image_path))[0]

        crop_save_path = os.path.join(OUTPUT_DIR, f"{base_name}_target{target_index}_crop.jpg")
        text_like_save_path = os.path.join(OUTPUT_DIR, f"{base_name}_target{target_index}_text_like.jpg")
        mask_save_path = os.path.join(OUTPUT_DIR, f"{base_name}_target{target_index}_lab_mask.jpg")

        cv2.imwrite(crop_save_path, crop_bgr)
        cv2.imwrite(text_like_save_path, cv2.cvtColor(text_like_rgb, cv2.COLOR_RGB2BGR))
        cv2.imwrite(mask_save_path, masked_crop_bgr)

        plt.figure(figsize=(14, 4))

        plt.subplot(1, 3, 1)
        plt.imshow(crop_rgb)
        plt.title("1. OCR検出領域の切り出し")
        plt.axis("off")

        plt.subplot(1, 3, 2)
        plt.imshow(text_like_rgb)
        plt.title("2. 文字っぽい画素")
        plt.axis("off")

        plt.subplot(1, 3, 3)
        plt.imshow(masked_crop_rgb)
        plt.title("3. Lab色でマスク後")
        plt.axis("off")

        plt.show()

        one_result = {
            "target_index": target_index,
            "text": text,
            "score": float(score),
            "box": box.tolist(),
            "crop_rect": crop_rect,
            "background_lab": background_lab.tolist(),
            "text_color_lab": text_color_lab.tolist(),
            "text_like_pixel_count": int(np.sum(text_like_mask > 0)),
            "masked_pixel_count": int(np.sum(color_mask > 0)),
            "crop_save_path": crop_save_path,
            "text_like_save_path": text_like_save_path,
            "mask_save_path": mask_save_path
        }

        image_result["results"].append(one_result)

    all_results.append(image_result)

# ============================================================
# 9. 最終サマリー
# ============================================================

print("\n" + "=" * 80)
print("最終サマリー")
print("=" * 80)

for image_result in all_results:
    print("\n画像:", image_result["image_name"])

    if len(image_result["results"]) == 0:
        print("  処理できた文字領域: なし")
        continue

    for r in image_result["results"]:
        print(
            f"  target_{r['target_index']} | "
            f"text={r['text']} | "
            f"score={r['score']:.3f} | "
            f"text_color_lab={np.round(r['text_color_lab'], 1).tolist()} | "
            f"masked_pixels={r['masked_pixel_count']}"
        )

print("\n結果画像の保存先:")
print(OUTPUT_DIR)
```
